1. What is hypothesis testing in statistics?
    - Hypothesis testing is a method used in statistics to make decisions or inferences about population parameters based on sample data. Basically, we test an assumption (hypothesis) to see if it's likely to be true or not.

2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
    - The null hypothesis (H‚ÇÄ) is the default assumption that there's no effect or no difference. The alternative hypothesis (H‚ÇÅ) is what we want to prove ‚Äî it states that there is an effect or a difference. We test the null hypothesis, and if the data strongly goes against it, we reject it in favor of the alternative.

3. What is the significance level in hypothesis testing, and why is it important?
    - The significance level (usually written as Œ±) is the threshold we set to decide whether to reject the null hypothesis. Common values are 0.05 or 0.01. It's important because it controls the chance of making a Type 1 error ‚Äî rejecting the null hypothesis when it's actually true.

4. What does a P-value represent in hypothesis testing?
    - The P-value shows the probability of getting a result as extreme as the one observed (or more) if the null hypothesis is true. It helps us decide whether to reject the null hypothesis.

5. How do you interpret the P-value in hypothesis testing?
    - If the P-value is less than the significance level (like 0.05), we reject the null hypothesis ‚Äî it means the result is statistically significant. If it's greater, we fail to reject the null ‚Äî meaning we don‚Äôt have enough evidence to say there‚Äôs a real effect.

6. What are Type 1 and Type 2 errors in hypothesis testing?
    - A Type 1 error happens when we reject a true null hypothesis (false positive), and a Type 2 error happens when we fail to reject a false null hypothesis (false negative). Type 1 is like thinking there's an effect when there isn‚Äôt, and Type 2 is missing an effect when there actually is one.

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
    - A one-tailed test checks for an effect in just one direction (greater than or less than), while a two-tailed test checks for any difference ‚Äî either higher or lower. Two-tailed is more common when we don‚Äôt know the direction of the effect in advance.

8. What is the Z-test, and when is it used in hypothesis testing?
    - The Z-test is used to test hypotheses about population means when the population standard deviation is known and the sample size is large. It's based on the normal distribution.

10. What is the T-distribution, and when should it be used instead of the normal distribution?
    - The T-distribution is similar to the normal distribution but has heavier tails. We use it when the sample size is small (typically n < 30) and the population standard deviation is unknown. It helps account for the extra uncertainty in small samples.

11. What is the difference between a Z-test and a T-test?
    - The main difference is that a Z-test is used when the population standard deviation is known and the sample size is large, while a T-test is used when the population standard deviation is unknown and the sample size is small. Z-test uses the normal distribution; T-test uses the t-distribution.

12. What is the T-test, and how is it used in hypothesis testing?
    - A T-test is used to compare the means of two groups and see if they are significantly different from each other. It‚Äôs especially useful when the sample size is small and the population standard deviation isn‚Äôt known.

13. What is the relationship between Z-test and T-test in hypothesis testing?
    - Both are used to test hypotheses about means. The Z-test is basically a special case of the T-test used when we have more information (like a known population standard deviation). As sample size increases, the T-distribution becomes more like the Z-distribution.

14. What is a confidence interval, and how is it used to interpret statistical results?
    - A confidence interval gives a range within which we expect the true population parameter to lie, with a certain level of confidence (like 95%). It helps us understand how precise our estimate is.

15. What is the margin of error, and how does it affect the confidence interval?
    - The margin of error is the amount we add and subtract from the sample statistic to create the confidence interval. A larger margin of error means a wider interval, which shows less precision.

16. How is Bayes' Theorem used in statistics, and what is its significance?
    - Bayes' Theorem is used to update the probability of a hypothesis based on new evidence. It‚Äôs important in situations where we want to revise beliefs as more data becomes available, like in medical testing or spam filtering.

17. What is the Chi-square distribution, and when is it used?
    - The Chi-square distribution is used in tests involving categorical data, especially for testing independence or goodness of fit. It‚Äôs skewed to the right and depends on degrees of freedom.

18. What is the Chi-square goodness of fit test, and how is it applied?
    - The Chi-square goodness of fit test checks if the observed frequencies match the expected frequencies for a single categorical variable. It helps us see if our data fits a particular distribution.

19. What is the F-distribution, and when is it used in hypothesis testing?
    - The F-distribution is used when comparing two variances or in ANOVA. It‚Äôs right-skewed and depends on two sets of degrees of freedom. It‚Äôs mainly used to test if multiple group means are equal.

20. What is an ANOVA test, and what are its assumptions?
    - ANOVA (Analysis of Variance) is used to compare the means of three or more groups. Its assumptions are: independence of observations, normality of the groups, and equal variances (homogeneity of variance).

21. What are the different types of ANOVA tests?
    - There are mainly three types:
    - One-way ANOVA: compares means of three or more groups based on one factor.
    - Two-way ANOVA: involves two factors and checks interaction between them.
    - Repeated Measures ANOVA: used when the same subjects are measured multiple times.

22. What is the F-test, and how does it relate to hypothesis testing?
    - The F-test compares two variances to check if they are significantly different. It's the basis for ANOVA and helps test if group means differ by looking at variance between and within groups.

1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
interpret the results?

```python
from scipy.stats import norm
import numpy as np

sample = [52, 50, 51, 53, 52, 50, 49, 54, 52, 51]
sample_mean = np.mean(sample)
population_mean = 50
std_dev = 2  # population standard deviation
n = len(sample)

z = (sample_mean - population_mean) / (std_dev / np.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z)))

print("Z-score:", z)
print("P-value:", p_value)

if p_value < 0.05:
    print("Reject null hypothesis.")
else:
    print("Fail to reject null hypothesis.")
```

3. Implement a one-sample Z-test using Python to compare the sample mean with the population mean?

```python
import numpy as np
from scipy.stats import norm

def one_sample_z_test(data, pop_mean, pop_std):
    sample_mean = np.mean(data)
    n = len(data)
    z = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
    p = 2 * (1 - norm.cdf(abs(z)))
    return z, p

sample = np.random.normal(102, 10, 40)
z_score, p_value = one_sample_z_test(sample, 100, 10)
print("Z-score:", z_score)
print("P-value:", p_value)
```

4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

sample_mean = 52
population_mean = 50
std_dev = 4
n = 30

z = (sample_mean - population_mean) / (std_dev / np.sqrt(n))
alpha = 0.05

x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

plt.plot(x, y)
plt.fill_between(x, y, where=(x < norm.ppf(alpha/2)) | (x > norm.ppf(1 - alpha/2)), color='red', alpha=0.3)
plt.axvline(z, color='blue', linestyle='--', label=f'Z = {z:.2f}')
plt.legend()
plt.title("Two-Tailed Z-Test Decision Region")
plt.show()
```

5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

alpha = 0.05
beta = 0.2
mean_null = 0
mean_alt = 1
std = 1

x = np.linspace(-4, 4, 1000)
null_dist = norm.pdf(x, mean_null, std)
alt_dist = norm.pdf(x, mean_alt, std)

z_crit = norm.ppf(1 - alpha)

plt.plot(x, null_dist, label="Null Hypothesis")
plt.plot(x, alt_dist, label="Alternative Hypothesis")
plt.fill_between(x, null_dist, where=(x > z_crit), color='red', alpha=0.3, label="Type I Error")
plt.fill_between(x, alt_dist, where=(x < z_crit), color='blue', alpha=0.3, label="Type II Error")
plt.legend()
plt.title("Type I and Type II Errors")
plt.show()
```

9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance?

```python
from scipy.stats import norm
import numpy as np

def confidence_interval(data, confidence=0.95):
    mean = np.mean(data)
    se = np.std(data, ddof=1) / np.sqrt(len(data))
    margin = norm.ppf((1 + confidence) / 2) * se
    return (mean - margin, mean + margin)

data = np.random.normal(100, 10, 40)
ci = confidence_interval(data)
print("Confidence Interval:", ci)
```

11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process?

```python
def bayes_theorem(prior_A, likelihood_B_given_A, prior_not_A, likelihood_B_given_not_A):
    numerator = likelihood_B_given_A * prior_A
    denominator = numerator + (likelihood_B_given_not_A * prior_not_A)
    posterior_A_given_B = numerator / denominator
    return posterior_A_given_B

# Example:
# Disease test example: P(Disease) = 0.01, P(Positive|Disease) = 0.99, P(Positive|No Disease) = 0.05
posterior = bayes_theorem(0.01, 0.99, 0.99, 0.05)
print(f"Posterior Probability of Disease given Positive Test: {posterior:.4f}")
```

13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed
data?

```python
import numpy as np

observed = np.array([[30, 10], [20, 40]])
row_totals = observed.sum(axis=1, keepdims=True)
col_totals = observed.sum(axis=0, keepdims=True)
grand_total = observed.sum()

expected = row_totals @ col_totals / grand_total
print("Expected Frequencies:\n", expected)
```
